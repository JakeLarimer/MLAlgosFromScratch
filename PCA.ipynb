Principal Component Analysis (or PCA) is an algorithm used to reduce the dimensionality of datasets. By dimensionality, we are referring to the number of features, not the number samples! The general goal of dimensionality reduction is to reduce the number of predictor variables in a dataset while preserving as much important information as possible. Various issues arrise when a dataset has a high dimensionality. Computation cost increases rapidly as dimensions increase, and ##more data points are required to sample the space##. Thus, the need for dimensionality reduction arises for certain datasets.

PCA essentially works by transforming the dataset into orthogonal $\textit{principal components}$, where the first principal component captures as much variance as possible, the second component explains as much of the remaining variance, and so on. The user can choose how many principal componenets to use, however something important to note is that if the number of components matches the number of features in a dataset, then the principal components together will explain all of the variance in a dataset. PCA then essentailly performs a rotation on the dataset.
