Principal Component Analysis (or PCA) is an algorithm used to reduce the dimensionality of datasets. By dimentionality, we are referring to the number of features, not the number samples! The general goal of dimensionality reduction is to reduce the number of predictor variables in a dataset while preserving as much important information as possible.
